```python
import matplotlib.pyplot as plt

plt.plot(df_random_out_1["time"], df_random_out_1["cost"], marker='o', label='SA', color='#e899dc')
plt.plot(df_neural_sa_1["time"], df_neural_sa_1["cost"], marker='o', label='Neural SA', color='#00BBC9')
plt.axhline(y=z, color='red', linestyle='--', alpha=0.8, linewidth=2, label=f'Óptimo TSPLIB ({z:.4f})')
plt.xlabel("Tiempo")
plt.ylabel("Costo")
plt.title("Comparación: SA vs Neural SA - M = 1")
plt.legend()
plt.grid(True)
plt.show()
```
